```python
df_train = pd.DataFrame()

df.columns = df.columns.str.replace('rainfall_train.', '', regex=False)

replacements = {'A': '2020', 'B': '2021', 'C': '2022'}

df_train['city'] = df['stn4contest']

df['ef_year'] = df['ef_year']
df['ef_year'] = df['ef_year'].replace(replacements).astype(int)

df_train['predict_time'] = pd.to_datetime(
    df['ef_year'].astype(str) + '-' +
    df['ef_month'].astype(str) + '-' +
    df['ef_day'].astype(str) + ' ' +
    df['ef_hour'].astype(str) + ':00:00'
)

df_train['0'] = 100 - df['v01']
df_train['1'] = df['v01'] - df['v02']
df_train['2'] = df['v02'] - df['v03']
df_train['3'] = df['v03'] - df['v04']
df_train['4'] = df['v04'] - df['v05']
df_train['5'] = df['v05'] - df['v06']
df_train['6'] = df['v06'] - df['v07']
df_train['7'] = df['v07'] - df['v08']
df_train['8'] = df['v08'] - df['v09']
df_train['9'] = df['v09']

df_train['dh'] = df['dh'].astype(int)
df_train['rainfall'] =  (df['class_interval'] > 0).astype(int)
df_train['class'] = df['class_interval'].astype(int)

df_train
```

```python
import datetime
from dateutil.relativedelta import relativedelta

df_train012 = df_train[df_train['city']=='STN012']

date = df_train012['predict_time'].iloc[0]
print(date)
end_date = df_train012['predict_time'].iloc[len(df_train012['predict_time'])-1]
print(end_date)
dates = []

while True:
  dates.append(date)
  if date == end_date:
    break

  if date.month == 10 and date.day == 10 and date.hour == 21:
    date = date + relativedelta(years=1)
    date = date.replace(month=5, day=1, hour=12, minute=0, second=0)
  else:
    date = date + pd.Timedelta(hours=3)

print(len(dates))
```
